## Connect X 

This is a notebook example of using Connect X.

* Author: https://github.com/ForeverHaibara 
* Game Intro: https://www.kaggle.com/c/connectx 

```python
from ConnectX import ConnectXBoard, ConnectXManager
from PIL import Image
from matplotlib import pyplot as plt

# all chessboards will be save in the path
path = r'D:\Python Projects\Chess\ConnectX\Figure.png' # customize your own path!
cntx_manager = ConnectXManager(path = path)
```

```python
def show_chess(path = path):
    fig = Image.open(path)
    plt.imshow(fig)
    #plt.axis('off')
    plt.show()
    plt.close()
```

```python
# create a chessboard
print( cntx_manager.create(['playerA','playerB'], turn = 0) )
show_chess()
```

```python
# one can create multiple chessboards at one time
print( cntx_manager.create(['playerC','playerD'], turn = 1) )
show_chess()
```

```python
# chessboards are stored in a list
print(cntx_manager.chesses)

# each player is assigned to the index of chessboard he or she is currently playing with
print(cntx_manager.playing)
```

```python
from time import sleep
from IPython.display import clear_output

# simulate a chess!
record = '1 1 1 1 5 2 2 3 4 3 3 4 6 7 4 1 2'
for i, x in enumerate(record.split()):
    player = ['playerA','playerB'][i % 2]
    clear_output()
    print( cntx_manager.play(player, x) )
    show_chess()
    sleep(0.5)
```

```python
# after one chess is done, it will be automatically removed from the chessboard list
print(cntx_manager.chesses)

# the corresponding players are also deleted
print(cntx_manager.playing)
```

```python
# playing with ConnectXBoard might be more convenient
# but remember to delete the chessboard after the chess is finished!

chessboard = ConnectXBoard()
record = '1 1 1 1 5 2 2 3 4 3 4 5 7 6 3 2 2'
for x in record.split():
    clear_output()
    print(chessboard.play(x)[0])
    chessboard.save_board(path)
    show_chess()
    sleep(0.5)

del chessboard
```
